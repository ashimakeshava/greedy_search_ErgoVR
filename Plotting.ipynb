```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

## Task 1: One color per row

```python
with open('collected_depths/task1.pickle', 'rb') as handle:
    depths = pickle.load(handle)
print(len(depths), "entries in file")
```

```python
num_moves = []
for d in depths:
    num_moves.append(d[1])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
plt.title('Task 1: Only one color per row')
bins= np.arange(min(num_moves),max(num_moves)+1,.5) #[5.,5.5,6.,6.5,7.,7.5,8.,8.5,9.,9.5,10.,10.5]
plt.hist(num_moves, bins,align='left')
#plt.title('Absolute frequencies of number of moves')
plt.xlabel('No. of moves until final state reached')
plt.ylabel('Absolute frequency')
plt.xticks(np.arange(min(num_moves),max(num_moves)+1))
plt.savefig('plots/task1')
plt.show()
```
